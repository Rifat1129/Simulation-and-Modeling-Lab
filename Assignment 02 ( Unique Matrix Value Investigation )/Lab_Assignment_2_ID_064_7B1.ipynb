```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
Task 1 : Create Your Personal Matrix
Let:
    ● d1 = last digit of your student ID
    ● d2 = second last digit of your student ID

    Create the matrix:
    A = ([[d1 + 2, d2 + 1], [2 * d1, d2 + 2]])

"""

#ID: 64

d1 = 4
d2 = 6

A = np.array([[d1 + 2, d2 + 1], [2 * d1, d2 + 2]])
print(A)
```

```python
"""
Task 2: Analyze Matrix A

Using NumPy, compute:
    ● Shape of A
    ● Determinant
    ● Rank
    ● Eigenvalues
    ● Inverse (only if determinant ≠ 0)

"""

# Shape
shape_A = A.shape

# Determinant
det_A = np.linalg.det(A)

# Rank
rank_A = np.linalg.matrix_rank(A)

# Eigenvalues
eig_A = np.linalg.eigvals(A)

# Inverse (only if determinant ≠ 0)
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Impossible"

print("Shape       :", shape_A)
print("Determinant :", det_A)
print("Rank        :", rank_A)
print("Eigenvalues :", eig_A)
print("Inverse     :\n", inv_A)
```

```python
"""
Task 3: Value Change Experiment

Create a new matrix B by changing exactly one value in A:
    ● Add +1 to one element of your choice
    ● You must clearly state which element you changed

"""

A = np.array([[d1 + 2, d2 + 1], [2 * d1, d2 + 2]])   # 1st element = d1 + 2
B = np.array([[d1 + 6, d2 + 1], [2 * d1, d2 + 2]])   # Change this element d1 + 2 = d1 + 6

print("Befor change Matrix A shown :\n",A)
print("\n After change Matrix B look like :\n",B)
```

```python
"""
Task 4: Re-analyze Matrix B

Compute the same properties as in Step 2.

"""

# Shape
shape_B = B.shape

# Determinant
det_B = np.linalg.det(B)

# Rank
rank_B = np.linalg.matrix_rank(B)

# Eigenvalues
eig_B = np.linalg.eigvals(B)

# Inverse (only if determinant ≠ 0)
if np.linalg.det(B) != 0:
    inv_B = np.linalg.inv(B)
else:
    inv_B = "Impossible"

print("Shape       :", shape_B)
print("Determinant :", det_B)
print("Rank        :", rank_B)
print("Eigenvalues :", eig_B)
print("Inverse     :\n", inv_B)
```
