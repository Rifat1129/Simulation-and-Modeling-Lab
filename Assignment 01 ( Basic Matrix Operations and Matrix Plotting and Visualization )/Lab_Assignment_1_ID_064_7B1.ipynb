```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""

Assignment 1: Basic Matrix Operations
1. Create two 3Ã—3 matrices with random integers in Python.
2. Perform addition, subtraction, and multiplication operations on these matrices.
3. Compute the determinant, inverse, and rank of one of the matrices.

"""


# Random Matrix
a = np.random.randint(0,10,size=(3,3))
b = np.random.randint(0,10,size=(3,3))

# Sum of 2 Random Matrix
sum = a + b

# Subtraction of 2 Random Matrix
sub = a - b

# Multiplication of 2 Random Matrix
mul = np.dot(a,b)

# Matrix Diterminent
det_a= np.linalg.det(a)
det_b = np.linalg.det(b)

# Matrix Inverse for a
if(np.linalg.det(a)!=0):
  inv_a = np.linalg.inv(a)
else:
  print("Matrix is Singular")

# Matrix Inverse for b
if(np.linalg.det(b)!=0):
  inv_b = np.linalg.inv(b)
else:
  print("Matrix is Singular")

# Rank of the Matrix
rank_a = np.linalg.matrix_rank(a)
rank_b = np.linalg.matrix_rank(b)

# Print Functions
print("Matrix A: \n",a,"\n")
print("Matrix B: \n",b,"\n")
print("Sum of Matrix: \n",sum,"\n")
print("Subtraction of Matrix: \n",sub,"\n")
print("Multiplication of Matrix: \n",mul,"\n")
print("Matrix Determinant of Mat A: \n",det_a,"\n")
print("Matrix Determinant of Mat B: \n",det_b,"\n")
print("Matrix Inverse of Mat A: \n",inv_a,"\n")
print("Matrix Inverse of Mat B: \n",inv_b,"\n")
print("Matrix Rank of Mat A: \n",rank_a,"\n")
print("Matrix Rank of Mat B: \n",rank_b,"\n")
```

```python
"""

Assignment 2: Matrix Plotting and Visualization
1. Generate and Plot Two Random Vectors:
a. Create two vectors, each containing 15 random floats of your choice.
b. Plot these vectors on the same graph using matplotlib to compare how they look.
c. Label the axes appropriately and give the plot a title.

"""

# Random Matrix with float
A = np.random.uniform(11,30, size=(5,3))
B = np.random.uniform(31,40, size=(5,3))

print("Matrix A: \n",A,"\n")
print("Matrix B: \n",B,"\n")

# Flattem 1D Array
flat_A = A.flatten()
flat_B = B.flatten()

print("Flattened Matrix A: \n",flat_A,"\n")
print("Flattened Matrix B: \n",flat_B,"\n")

# Create an Index for x-Axis
x = np.arange(len(flat_A))

# Line Plot
plt.plot(x, flat_A, marker = "o", linestyle = "-", label="Matrix A")
plt.plot(x, flat_B, marker = "s", linestyle = "--", label="Matrix B")

# Label and Title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Float Flatten Matrices A and B")
plt.legend()
plt.grid(True)

# Show the Plot
plt.show()
```
